```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#fetching data
data = fetch_california_housing() 
df = pd.DataFrame(data.data, columns=data.feature_names) 
df['target'] = data.target

#plotting scatter plot of features
selected_features = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms']
plt.figure(figsize=(16, 12))
for feature in selected_features:
    plt.scatter(df[feature], df['target'], label=feature)
plt.xlabel("Feature Values")
plt.ylabel("Median House Value")
plt.title("Scatter Plot of California Housing Dataset Features")
plt.legend()
plt.show()


#splitting data into train and test
X = df[selected_features]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=32)

#create model
model = LinearRegression()
model.fit(X_train, y_train)


#predicting on test data
y_pred = model.predict(X_test)

#mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

#Suggest measures to improve mean squared error.
#Answer: We can use feature selection techniques to select the best features for the model.
#We can also use regularization techniques like Lasso and Ridge regression to reduce overfitting.
#we can also apply L1 or L2 regularization to the linear regression model.
#We can also use cross validation to reduce overfitting.
```
