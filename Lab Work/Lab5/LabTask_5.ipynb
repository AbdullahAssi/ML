```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

data = pd.read_csv("data.csv")


print(data.head())

X = data.iloc[:, 1:].values

kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(X)

# Visualize the clustering results
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='cividis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='green', marker='*', label='Centroids')
plt.title("K-Means Clustering Results")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

data = pd.read_csv("data.csv")

print(data.head())


X = data.iloc[:, 1:].values

agg_clustering = AgglomerativeClustering(n_clusters=6)
y_pred = agg_clustering.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.title("Hierarchical Clustering Results")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("data.csv")

print(data.head())

X = data.iloc[:, 1:].values
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.5, min_samples=5)
y_pred = dbscan.fit_predict(X_standardized)

plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='plasma')
plt.title("DBSCAN Clustering Results")
plt.show()
```
