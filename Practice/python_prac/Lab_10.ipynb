```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('HistoricalPrices.csv')
data.sample(5)
```

```python

fig, axs = plt.subplots(len(data.columns) - 1, 1, figsize=(10, 10))

for i, column in enumerate(data.columns[1:]):
    axs[i].plot(data['Date'], data[column])
    axs[i].set_xlabel('Date')
    axs[i].set_ylabel(column)
    axs[i].set_title(column)
    
plt.tight_layout()

plt.show()
```

```python

data['Date'] = pd.to_datetime(data['Date'])
colors = ['red', 'blue', 'green', 'orange']

fig, ax = plt.subplots()

for i, column in enumerate(data.columns[1:]):
    ax.plot(data['Date'], data[column], color=colors[i], label=column)

ax.set_xlabel('Date')
c:\Users\Abdullah\Downloads\Historical_Prices.csv
ax.set_ylabel('Value')

ax.set_title('Multiple Line Plot')

# Add a legend
ax.legend()

plt.show()
```

```python

data['Date'] = pd.to_datetime(data['Date'])

data['Month'] = data['Date'].dt.month

print(data['Month'])
# print(data.columns)

datagroup = data.groupby('Month')[' Close'].mean()

datagroup.plot(kind='bar', xlabel='Month', ylabel='Average Close', title='Average Monthly Close Prices')

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(ticks=range(1, 13), labels=month_names, rotation=0)

plt.show()
```

```python

plt.scatter(data[' Open'], data[' Close'])
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Scatter Plot: Open vs Close')
plt.show()
```
