```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


data = pd.read_csv('sales_data_sample.csv' , encoding='latin1')
data.sample(5)
```

```python
data.columns
```

```python
data['SALES'].value_counts()
```

```python
data['SALES'].describe()
```

```python

sales_array = np.array(data['SALES'])
print(sales_array)
```

```python
sales_by_country = data.groupby('COUNTRY')['SALES'].sum()
print(sales_by_country)
```

```python
average_sales_by_product = data.groupby('PRODUCTLINE')['SALES'].mean()
print(average_sales_by_product)
```

```python

sales_by_country.plot(kind='bar' , figsize=(10,6))

plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.title('Total Sales by Country') 

plt.show()
```

```python

average_sales_by_product.plot(kind='pie', figsize=(10, 6), autopct='%1.1f%%')

plt.title('Average Sales by Product')

plt.show()
```

bar chart to show the distribution avarage sales across productline

```python
average_sales_by_product.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Product Line')
plt.ylabel('Average Sales')
plt.title('Average Sales by Product Line')

plt.show()
```
