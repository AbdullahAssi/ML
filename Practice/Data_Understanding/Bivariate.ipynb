```python
import pandas as pd
import numpy as np
import seaborn as sns   

tips = sns.load_dataset('tips')
titanic = pd.read_csv('train.csv') 
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris') 
```

<h1>Scatter Plot : Numerical Numerical</h1>

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips)
sns.relplot(
    data=tips, x="total_bill", y="tip",
    col="time", hue="day", style="day",
    kind="scatter"
)
```

<h1> Bar Plot : Nmerical - Catagorical </h1>

```python

sns.barplot(x='Pclass', y='Fare', hue='Sex', data=titanic)
```

<h1> Box Plot : Nmerical - Catagorical </h1>

```python
sns.boxenplot(x='Sex', y='Age', hue='Survived', data=titanic)
```

<h1>Distplot Numerical -Catagorical</h1>

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False ) 
```

<h1>Heatmap Catagorical Catagorical</h1>

```python
sns.heatmap(pd.crosstab(titanic['Survived'], titanic['Pclass']), annot=True, fmt='d')
```

```python
survival_rate = titanic.groupby('Embarked')['Survived'].mean() * 100
print(survival_rate)


survival_rate = titanic.groupby('Sex')['Survived'].mean() * 100
print(survival_rate)
```

<h1>ClusterMap (Catagorical - Catagorical)</h1>

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Pclass'], margins=True))
```

<h1>Pair Plot </h1>

```python
sns.pairplot(iris,hue='species')
```
